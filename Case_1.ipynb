```python
def latin_to_kaz(txt):
  alphabet = {'A' : 'a', 'Á' : 'ә', 'B': 'б', 'V' : 'в', 'G':'г', 'Ǵ':'ғ', 'D':'д', 'E':"е", 'ZH':'ж', 'Z':'з', 'I':'й', 'K':'к', 'Q':'қ', 'L':'л', 'M':'м', 'N': 'н', 'Ń':'ң', 'O': 'о', 'Ö':'ө','P':'п', 'R':'р', 'S':'с', 'T':'т', 'Ý':'у', 'Ū':'ұ', 'Ü':'ү', 'F':'ф', 'H':'х', 'TS':'ц', 'CH':'ч', 'SH':'ш', 'Y':'ы', 'I':'і', 'IU':'ю', 'IA':'я', 'I':'и', 'Ú': 'ү', 'U': 'ұ', 'Ó': 'ө'}
  txt_new = []
  txt = txt.upper()
  txt = txt + '.'
  for i in range(len(txt)):
    if txt[i] == 'S' and txt[i-1] == 'T':
      txt_new.pop()
      txt_new.append(alphabet['TS'])
    elif txt[i] == 'H' and txt[i-1] == 'Z':
      txt_new.pop()
      txt_new.append(alphabet['ZH'])
    elif txt[i] == 'H' and txt[i-1] == 'C':
      txt_new.pop()
      txt_new.append(alphabet['CH'])
    elif txt[i] == 'H' and txt[i-1] == 'S':
      txt_new.pop()
      txt_new.append(alphabet['SH'])
    elif txt[i] == 'U' and txt[i-1] == 'I':
      txt_new.pop()
      txt_new.append(alphabet['IU'])
    elif txt[i] == 'A' and txt[i-1] == 'I':
      txt_new.pop()
      txt_new.append(alphabet['IA'])
    else:
      if txt[i] in alphabet:
        if txt[i] == 'I' and (txt[i-1] == 'A' and txt[i-1] == 'Á' and txt[i-1] == 'Ö' and txt[i-1] == 'E' and txt[i-1] == 'O' and txt[i-1] == 'Ý' and txt[i-1] == 'Ū' and txt[i-1] == 'Ü' and txt[i-1] == 'Y'):
          txt_new.append('й')
        else:
          txt_new.append(alphabet[txt[i]])
      else:
        txt_new.append(txt[i])
  return ''.join(txt_new)
```

##**BM25**

```python
import math

class BM25:
    def __init__(self, k1=1.5, b=0.75):
        self.k1 = k1
        self.b = b
        self.doc_lengths = None
        self.avg_doc_length = None
        self.doc_term_freqs = None
        self.doc_count = None
        self.idf = None

    def fit(self, corpus):
        # Calculate document lengths and term frequencies
        self.doc_lengths = [len(doc) for doc in corpus]
        self.avg_doc_length = sum(self.doc_lengths) / len(corpus)
        self.doc_term_freqs = []
        self.doc_count = len(corpus)

        for doc in corpus:
            term_freqs = {}
            for term in doc:
                term_freqs[term] = term_freqs.get(term, 0) + 1
            self.doc_term_freqs.append(term_freqs)

        # Calculate IDF for all unique terms in the corpus
        self.idf = {}
        for doc in self.doc_term_freqs:
            for term in doc:
                self.idf[term] = self.idf.get(term, 0) + 1

        for term, doc_count in self.idf.items():
            self.idf[term] = math.log((self.doc_count - doc_count + 0.5) / (doc_count + 0.5) + 1.0)

    def score(self, query, doc):
        score = 0
        doc_length = self.doc_lengths[doc]

        for term in query:
            if term in self.doc_term_freqs[doc]:
                idf = self.idf[term]
                tf = self.doc_term_freqs[doc][term]
                numerator = tf * (self.k1 + 1)
                denominator = tf + self.k1 * (1 - self.b + self.b * doc_length / self.avg_doc_length)
                score += idf * numerator / denominator

        return score

    def rank(self, query, corpus):
        scores = []

        for doc in range(len(corpus)):
            scores.append(self.score(query, doc))

        # Sort the documents by BM25 score in descending order
        ranked_docs = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)

        return ranked_docs
```

##**TF - IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# fit and transform the DataFrame column
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])
```

```python
indices = []
for question in questions['question']:
  # transform the user's question
  question = re.sub(r'[^\w\s]', '', question).lower()
  user_question_vector = tfidf_vectorizer.transform([question])

  # calculate cosine similarities between the user's question and all texts
  cosine_similarities = linear_kernel(user_question_vector, tfidf_matrix)

  # get the index of the most similar text
  most_similar_index = cosine_similarities.argmax()

  print(f"The question: {question}")
  print(f"Most relevant text: {df.iloc[most_similar_index]['text']}")
  print(f"Index of the most relevant text: {most_similar_index}")
  print(f"Ind (to return) of the most relevant text: {df.iloc[most_similar_index]['index']}\n\n")
  indices.append(df.iloc[most_similar_index]['index'])
```

```python
dt = pd.DataFrame()
dt['id'] = questions['id']
dt['index'] = indices
dt.to_csv('out-4.csv', index = False)
```
